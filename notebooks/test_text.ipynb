```python
def tester_mon_texte(texte_brut):
    """
    Applique le pipeline de nettoyage complet.
    """
    texte_corrige = english_cleaners(texte_brut)
    
    print("-" * 30)
    print(f"TEXTE D'ENTRÉE : {texte_brut}")
    print(f"TEXTE CORRIGÉ : {texte_corrige}")
    print("-" * 30)
    
    if any(char.isdigit() for char in texte_brut) and not any(char.isdigit() for char in texte_corrige):
        print("Info : Les nombres ont été correctement convertis en lettres.")
    
    return texte_corrige
```
